```python
trades['Timestamp IST'] = pd.to_datetime(
    trades['Timestamp IST'],
    dayfirst=True
)

trades['date'] = trades['Timestamp IST'].dt.date
trades['date'] = pd.to_datetime(trades['date'])

print(trades[['Timestamp IST', 'date']].head())
```

```python
merged = pd.merge(
    trades,
    fear[['date', 'classification']],
    on='date',
    how='left'
)

print(merged.shape)
```

```python
merged[['date', 'classification']].head(20)
```

```python
merged.groupby('classification')['Closed PnL'].mean()
```

```python
merged.groupby('classification')['Closed PnL'].sum()
```

```python
merged['win'] = merged['Closed PnL'] > 0

merged.groupby('classification')['win'].mean() * 100
```

```python
import matplotlib.pyplot as plt

avg_pnl = merged.groupby('classification')['Closed PnL'].mean()

plt.figure(figsize=(8,5))
avg_pnl.plot(kind='bar')

plt.title('Average Profit by Market Sentiment')
plt.ylabel('Average Closed PnL')
plt.xlabel('Market Sentiment')

plt.tight_layout()

plt.savefig('avg_profit_by_sentiment.png')

plt.show()
```

```python
win_rate = merged.groupby('classification')['win'].mean()*100

plt.figure(figsize=(8,5))
win_rate.plot(kind='bar')

plt.title('Win Rate by Market Sentiment')
plt.ylabel('Win Rate (%)')
plt.xlabel('Market Sentiment')

plt.tight_layout()

plt.savefig('win_rate_by_sentiment.png')

plt.show()
```

```python
total_pnl = merged.groupby('classification')['Closed PnL'].sum()

plt.figure(figsize=(8,5))
total_pnl.plot(kind='bar')

plt.title('Total Profit by Market Sentiment')
plt.ylabel('Total Closed PnL')
plt.xlabel('Market Sentiment')

plt.tight_layout()

plt.savefig('total_profit_by_sentiment.png')

plt.show()
```
